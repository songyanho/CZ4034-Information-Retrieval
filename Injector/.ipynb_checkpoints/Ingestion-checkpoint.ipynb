```python
import pandas as pd
from elasticsearch import Elasticsearch
from elasticsearch import helpers
from textblob import TextBlob, Word
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
```

```python
# Default configuration
stopwords = set(stopwords.words('english'))
stemmer = PorterStemmer()
```

```python
def text_preprocessing(raw_review):
    try:
        review = TextBlob(raw_review.decode('ascii', 'ignore').encode('ascii','ignore'))
        # Tokenization
        # Normalization
        # Stemming
        # Remove stopwords
        clean_tokens = []
        for word in review.words:
            stem = stemmer.stem(word.lower())
            if stem not in stopwords:
                clean_tokens.append(stem)

        return ' '.join(clean_tokens)
    except:
        return ''
    
def sentiment_preprocessing(raw_review):
    try:
        review = TextBlob(raw_review.decode('ascii', 'ignore').encode('ascii','ignore'))
        return review.sentiment
    except:
        return 0
```

```python
# Index to elasticsearch
es = Elasticsearch(['http://elastic:changeme@localhost:9200/'])
```

```python
file_count = 0

for reviews in pd.read_csv('/Volumes/Seagate Backup Plus Drive/reviews_2017.csv', names=[
    'name',
    'url',
    'year',
    'poster',
    'meterClass',
    'meterScore',
    'casts',
    'user',
    'review',
    'superReviewer',
    'reviewScore'
], chunksize=1000):
    
    reviews.fillna(0, inplace=True)
    
    file_count += 1

    # Clean text
    reviews['tokens'] = reviews['review'].apply(text_preprocessing)
    reviews['sentiment'] = reviews['review'].apply(sentiment_preprocessing)
    actions = [{
    "_index": "review",
    "_type": "rottentomato",
    "_source": review.to_dict()} for _, review in reviews.iterrows()]
    
    try:
        helpers.bulk(es, actions)
    except Exception as e:
        print(e)
    print(file_count)
#     reviews.to_csv('/Volumes/Seagate Backup Plus Drive/chunks/reviews_{}.csv'.format(file_count))
```
